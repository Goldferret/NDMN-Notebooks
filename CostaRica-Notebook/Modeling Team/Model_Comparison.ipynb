# Random state 30

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load models
with open('ML_Models/rf_model.pkl', 'rb') as f:
    rf_model = pickle.load(f)

with open('ML_Models/svm_model_2.pkl', 'rb') as f:
    svm_model = pickle.load(f)

models = {
    'Random Forest': rf_model,
    'SVM': svm_model
}

# Load the dataset
df = pd.read_parquet('../Costa_Rica_Data/Data Acquisition Output/extracted_gee_data/en_clean_gee_data.parquet')

# Set features and target variable
X = df[['ndvi', 'evi', 'savi', 'brightness', 'wetness', 'slope', 'aspect', 'mTPI', 'ocs_1mMed', 'sand_1mMed', 'topDiv']]
y = df['Use']

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)

# Evaluate model
def evaluate_model(model, X_test, y_test):
    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    # Create confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    return accuracy, precision, recall, f1, cm

# Store the evaluation results
evaluation_results = {}

# Evaluate both models
for model_name, model in models.items():
    accuracy, precision, recall, f1, cm = evaluate_model(model, X_test, y_test)
    evaluation_results[model_name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Confusion Matrix': cm
    }

# Print evaluation results
for model_name, results in evaluation_results.items():
    print(f"\n{model_name} Evaluation:")
    print(f"Accuracy: {results['Accuracy']:.2f}")
    print(f"Precision: {results['Precision']:.2f}")
    print(f"Recall: {results['Recall']:.2f}")
    print(f"F1-Score: {results['F1-Score']:.2f}")

    # Plot confusion matrix
    plt.figure(figsize=(6, 5))
    sns.heatmap(results['Confusion Matrix'], annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
```
